```python
r = requests.get("https://bodacc-datadila.opendatasoft.com/api/records/1.0/search/?dataset=annonces-commerciales&q=&sort=dateparution&facet=publicationavis&facet=publicationavis_facette&facet=typeavis&facet=typeavis_lib&facet=familleavis&facet=familleavis_lib&facet=numerodepartement&facet=departement_nom_officiel")
```

```python
r.content
```

```python
r.json()
```

```python
r.json()['records']
```

```python
d = {'publicationavis': [], 'listepersonnes': [],'registre':[],'ispdf_unitaire':[],'cp':[],'jugement':[],'commercant':[],'dateparution':[],'tribunal':[], 'typeavis':[],'familleavis_lib':[]}
```

```python
for i,_ in enumerate(r.json()['records']):
    d['publicationavis'].append(r.json()['records'][i]['fields']['publicationavis'])
    d['listepersonnes'].append(r.json()['records'][i]['fields']['listepersonnes'])
    d['registre'].append(r.json()['records'][i]['fields']['registre'])
    d['cp'].append(r.json()['records'][i]['fields']['cp'])
    d['ispdf_unitaire'].append(r.json()['records'][i]['fields']['ispdf_unitaire'])
    d['jugement'].append(r.json()['records'][i]['fields']['jugement'])
    d['commercant'].append(r.json()['records'][i]['fields']['commercant'])
    d['dateparution'].append( r.json()['records'][i]['fields']['dateparution'])
    d['tribunal'].append(r.json()['records'][i]['fields']['tribunal'])
    d['typeavis'].append(r.json()['records'][i]['fields']['typeavis'])
    d['familleavis_lib'].append(r.json()['records'][i]['fields']['familleavis_lib'])
```

```python
d
```

```python
df_bodacc = pd.DataFrame.from_dict(d)
df_bodacc
```
